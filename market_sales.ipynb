```python
plt.scatter(df["Unit price"],df["cogs"])
plt.title("Units")
plt.xlabel("price")
plt.ylabel("cogs")
```

```python
plt.hist(df["Quantity"],bins=5,edgecolor="red")
plt.title("Quantity distibution")
plt.xlabel("value")
plt.ylabel("frequency")
```

```python
plt.hist([df["Total"],df["Income"]],bins=8,edgecolor="red",label=["Total","Income"])
plt.title("distribution Total and Income")
plt.xlabel("amt")
plt.ylabel("frequency")
plt.legend()
plt.show()
```

```python
gender=df["Gender"].value_counts()
print(gender)
plt.figure(figsize=(7,7))
plt.pie(gender,labels=["female","Male"],autopct="%1.0f%%",startangle=140,colors=["green","red"],wedgeprops={"width":0.2}) #wedge for givig domut chart
plt.title("Male and Female")
```

```python
sns.histplot(data=df,x="Total")#data=df is common syntax
plt.title('Histogram Plot based on Total')
plt.show()
```

```python
sns.scatterplot(data=df,x="Total",y="Rating")
plt.xlabel("Total")
plt.ylabel("Rating")
plt.title('Scatter Plot based on Rating and Total')
plt.show()
```

```python
sns.barplot(data=df,x="Gender",y="Unit price") #x axis should be categorical
plt.title('BarPlot based on Gender and Unit Price')
plt.xlabel("gender")
plt.ylabel("price")
plt.show()
```

```python
sns.lineplot(x="Rating",y="Total",data=df)
plt.title('Line Plot based on Rating and Total')
plt.xlabel("rating")
plt.ylabel("total ")
plt.show()
```

```python
sns.swarmplot(x="Gender",y="Total",data=df)
plt.title('SwarmPlot based on Gender and Total')
plt.show()
```

```python
sns.swarmplot(data=df,x="Total",y="Gender",hue="Quantity",orient="h")  
sns.set_theme(style="darkgrid")# or sns.set_theme(style="whitegrid")# to give background theme
plt.title('Horizontal SwarmPlot ')
plt.show()
```

```python
sns.boxplot(data=df,x="Gender",y="Total")
plt.title('Box Plot based on Gender and Total ')
```

```python
sns.violinplot(data=df,x="Gender",y="Total",hue="Customer type")
plt.title('Violin Plot based on Gender and Total ')
```

```python
sns.relplot(data=df,x="Total",y="Rating",hue="Customer type")
plt.title('Relational Plot based on Total and Rating')
```

```python
sns.kdeplot(data=df,x="Total",y="Rating",shade=True)
plt.title('Heat Map based on Total and Rating')
```
